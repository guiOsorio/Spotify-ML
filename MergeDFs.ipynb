```python
import pandas as pd
```

```python
# Get data frames
df_gui = pd.read_csv("spotify_mytracks_data.csv") 
df_soly = pd.read_csv("spotify_solytracks_data.csv") 

df_gui.head()
```

```python
# Merge dfs
df = df_gui.merge(df_soly, how='outer')

print('Number of total columns: ', df_gui.shape[0] + df_soly.shape[0])
print(df.shape)
df.head()
```

```python
# Group the entries by artist_name and track_name
grouped = df.groupby(['artist_name','track_name'], as_index=True).size()

# Check for duplicates
grouped[grouped > 1].count() # 59 duplicates
```

```python
# Songs shared by both datasets
grouped[grouped > 1]
```

```python
# Drop duplicates
df.drop_duplicates(subset=['artist_name','track_name'], inplace=True)
```

```python
# Create csv file with data retrieved
df.to_csv('spotify_initialdata.csv')
```
